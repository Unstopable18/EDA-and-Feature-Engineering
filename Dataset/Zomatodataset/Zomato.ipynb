```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on="Country Code", how="left")
final_df.head()
```

```python
final_df.columns
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_count=final_df.Country.value_counts().values
country_count
```

```python
plt.pie(labels=country_name[0:],x=country_count[0:], autopct='%1.2f%%')
```

Observation: 90.59% of transaction are from India as comapred to other 14 countries followed by india there is United States with 4.54% of transaction, therafter comes United Kingdom holding 0.84% transaction and remaining 12 countries combining hold 4.03% transaction.

```python
rating_df=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating_df
```

Observation: <br>
Rating--------Color--------Conclusion<br>

0.1--2.4|---|White     |---|Not Rated<br>
0.1--2.4|---|Red       |---|Poor<br>
2.4--3.4|---|Orange    |---|Average<br>
3.4--3.9|---|Yellow    |---|Good<br>
3.9--4.4|---|Green     |---|Very Good<br>
4.4--4.9|---|Dark Green|---|Excellent

```python
sns.barplot(x="Aggregate rating",y="Rating Count", data=rating_df)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count", hue="Rating color", palette=["Black","Red","Orange","Yellow","Green","Blue"], data=rating_df)
```

Observation:
Not rated is way more then other rated once
Most of the ratings can be seen in range of 2.8 - 3.8


```python
sns.countplot(x="Rating color", palette=["Black","Red","Orange","Yellow","Green","Blue"], data=rating_df)
```

Observation:
Most of the rating records are in orange followed by red 
Also can see yellow green and dark green are mostly same, can say there is not much difference
